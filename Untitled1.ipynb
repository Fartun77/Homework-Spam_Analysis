```python
#1 

import matplotlib.pyplot as plt

# Sort values to make the line chart more interpretable
df_sorted = df.sort_values(by='capitalLong')

plt.figure(figsize=(10, 6))
plt.plot(df_sorted['capitalLong'].values)
plt.title('Distribution of CapitalLong')
plt.xlabel('Index')
plt.ylabel('CapitalLong')
plt.show()
```

```python
#2  
import matplotlib.pyplot as plt

# Count occurrences of each category in the 'type' column
type_counts = df['type'].value_counts()

plt.figure(figsize=(8, 6))
type_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Frequency of Spam and Non-Spam Emails')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Keep labels horizontal
plt.show()

```

```python
#3
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['num3d'], df['num000'], alpha=0.5)
plt.title('Relationship between num3d and num000')
plt.xlabel('num3d')
plt.ylabel('num000')
plt.show()

```

```python
#4

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['money'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Money Feature')
plt.xlabel('Money')
plt.ylabel('Frequency')
plt.show()

```

```python
#5

import matplotlib.pyplot as plt

# Count occurrences of each category in the 'type' column
type_counts = df['type'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])
plt.title('Proportion of Spam and Non-Spam Emails')
plt.show()

```

```python
#16

import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Generate the correlation matrix
correlation_matrix = numeric_df.corr()

# Set up the plot
plt.figure(figsize=(15, 10))

# Plot the heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
#17
# Select only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix and unstack it to make it easier to sort
correlation_matrix = numeric_df.corr().unstack()

# Sort correlations by absolute value, in descending order
sorted_correlations = correlation_matrix.sort_values(key=abs, ascending=False)

# Remove self-correlations (correlation of a feature with itself)
sorted_correlations = sorted_correlations[sorted_correlations != 1.0]

# Display the two highest positive and two highest negative correlations
highest_positive = sorted_correlations.head(2)
highest_negative = sorted_correlations.tail(2)

print("Highest Positive Correlations:")
print(highest_positive)
print("\nHighest Negative Correlations:")
print(highest_negative)

```

```python
## 23 A bar chart is perfect for representing categorical data, facilitating straightforward comparison of amounts between different categories (e.g., counts of spam and non-spam emails).
```

```python
##24 A scatter plot is suitable when exploring relationships between two continuous numerical variables, helping to identify correlations, patterns, or outliers.
```

```python
##25 Benefits: Easy to depict the fractions of categories in relation to a total and the disadvantge is that it is difficult to understand with numerous categories or slight variations.
```

```python
##26 line chart is better suited for displaying trends over time or a continuous series, while a histogram is more effective for illustrating the distribution of one variable.
```

```python
##27 A table offers more information when precise values are required across various categories or dimensions, like in a financial report displaying comprehensive statistics.
```

```python
##28 A histogram illustrates the distribution of a variable, indicating frequency, dispersion, form, and the existence of any outliers or clusters.
```
